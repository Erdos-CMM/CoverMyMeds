# Some visualization 

In this notebook I make a few more plots to explore the relevance of different features. 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_theme(style='whitegrid')
```

```python
data = pd.read_csv('../data/processed_data/processed_data.csv')
```

```python
data.head()
```

```python
data.columns
```

## Count plots

Which bayer approves/rejects the most?

Below is the count plot of payers (the "bin" column).

```python
plt.figure(figsize=(10,5),dpi=200)
p = sns.countplot(x='bin',data=data,hue='pharmacy_claim_approved')

p.axes.set_title("claim approval by bin",fontsize=24)
p.set_xlabel("bin",fontsize=20)
p.set_ylabel("Count",fontsize=20)
p.tick_params(labelsize=14)
plt.show()
```

Only the payer:999001 approves most of the claims. All others reject more than accept. The payer:417614 rejects the most claims. 

```python
plt.figure(figsize=(10,5),dpi=200)
p = sns.countplot(x='day_of_week',data=data,hue='pharmacy_claim_approved')

p.axes.set_title("claim approval by day of week",fontsize=24)
p.set_xlabel("day of week",fontsize=20)
p.set_ylabel("Count",fontsize=20)
p.tick_params(labelsize=14)
plt.show()
```

Counting approval and denial of PA by each payer.

```python
plt.figure(figsize=(10,5),dpi=200)
p = sns.countplot(x='bin',data=data,hue='pa_approved')

p.axes.set_title("PA approval by bin",fontsize=24)
p.set_xlabel("bin",fontsize=20)
p.set_ylabel("Count",fontsize=20)
p.tick_params(labelsize=14)
plt.show()
```

All payers approves most of the PAs.

Count plot of reject codes.

```python
plt.figure(figsize=(10,5),dpi=200)
p = sns.countplot(x='reject_code',data=data,hue='pa_approved') 

p.axes.set_title("PA approval by reject code",fontsize=24)
p.set_xlabel("reject code",fontsize=20)
p.set_ylabel("Count",fontsize=20)
p.tick_params(labelsize=14)
plt.show()
```

### How does the correct diagnosis affects the approval of PA? 

```python
plt.figure(figsize=(10,5),dpi=200)
p = sns.countplot(x='correct_diagnosis',data=data,hue='pa_approved')

p.axes.set_title("PA approval by correct diagnosis",fontsize=24)
p.set_xlabel("correct diagnosis",fontsize=20)
p.set_ylabel("Count",fontsize=20)
p.tick_params(labelsize=14)
plt.show()
```

It seems the correct diagnosis has no significant impact on PA approval.
