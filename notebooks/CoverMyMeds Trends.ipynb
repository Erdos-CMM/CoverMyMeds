Some notes on rejection codes: 

There are 3 rejection codes that we focus on:
70, 75, and 76. 

A code 70 implies that a drug is not covered by the plan and is not on 
formulary, and typically implies that another course of therapy should be pursued. A
code 75 implies that a drug is on the formulary but does not have preferred status and
requires a prior authorization (PA). A PA is a form that care providers submit on behalf
of their patient making the case that the therapy as described is a critical course of
treatment. A code 76 simply means that the drug is covered, but that the plan
limitations have been exceeded, which means that the limitations on the number of fills
for that medication has been met. We might expect there to be variation in the type of
reject codes we see for certain drugs by the payer.

```python
# set the figure size
plt.figure(figsize=(7, 7))

# from raw value to percentage
code70 = data[data.reject_code==70].groupby('drug')['reject_code'].count().reset_index()
code75 = data[data.reject_code==75].groupby('drug')['reject_code'].count().reset_index()
code76 = data[data.reject_code==76].groupby('drug')['reject_code'].count().reset_index()
#code0 = data[data.reject_code==0.].groupby('drug')['reject_code'].count().reset_index()                                   
total = data[data.reject_code!=0.].groupby('drug')['reject_code'].count().reset_index()
             
             
             

code70['percentage_reject_code'] = [i / j * 100 for i,j in zip(code70['reject_code'], total['reject_code'])]
code75['percentage_reject_code'] = [i / j * 100 for i,j in zip(code75['reject_code'], total['reject_code'])]
code76['percentage_reject_code'] = [i / j * 100 for i,j in zip(code76['reject_code'], total['reject_code'])]
#total['total_bill'] = [i / j * 100 for i,j in zip(total['total_bill'], total['total_bill'])]

# bar chart 1 -> top bars (group of 'smoker=No')
bar1 = sns.barplot(x="drug",  y="percentage_reject_code", data=code70, color='darkblue', label='code 70')

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar2 = sns.barplot(x="drug",  y="percentage_reject_code", data=code75, color='lightblue', label='code 75')

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar3 = sns.barplot(x="drug",  y="percentage_reject_code", data=code76, color='grey', label='code 76')

# bar chart 2 -> bottom bars (group of 'smoker=Yes')
#bar4 = sns.barplot(x="drug",  y="reject_code", data=code0, color='darkgreen', label='code 0')

# add legend
#top_bar = mpatches.Patch(color='darkblue', label='smoker = No')
#bottom_bar = mpatches.Patch(color='lightblue', label='smoker = Yes')
#plt.legend(handles=[top_bar, bottom_bar])

plt.legend()

plt.title('% of reject codes by drug type')
# show the graph
plt.show()
```
