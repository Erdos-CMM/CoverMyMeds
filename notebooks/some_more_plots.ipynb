# Some visualization 

In this notebook I make a few more plots to explore the relevance of different features. 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_theme(style='whitegrid')
```

```python
data = pd.read_csv('../data/processed_data/processed_data.csv')
```

```python
data.head()
```

```python
data.columns
```

## Count plots

Which bayer approves/rejects the most?

Below is the count plot of payers (the "bin" column).

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='bin',data=data,hue='pharmacy_claim_approved')
```

Only the payer:999001 approves most of the claims. All others reject more than accept. The payer:417614 rejects the most claims. 

Counting approval and denial of PA by each payer.

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='bin',data=data,hue='pa_approved')
```

All payers approves most of the PAs. 

Counting type of drugs and claim approval. 

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='drug',data=data,hue='pharmacy_claim_approved')
```

Count plot of reject codes.

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='reject_code',data=data,hue='pa_approved')
```

The reject code 70 has the hight probability of PA denied. 

### How does the correct diagnosis affects the approval of PA? 

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='correct_diagnosis',data=data,hue='pa_approved')
```

It seems the correct diagnosis has no significant impact on PA approval.
